## 5) 워드 클라우드

① 인터넷에서 좋은 글(영문으로 된 연설문, 혹은 기사, 책 내용)을 찾아, txt로 저장하고 워드 클라우드를 만들어 봅시다.

* 코드는 예제파일을 참조하세요.
* 불필요한 단어/조사는 제거하세요.
* 글의 내용을 한눈에 파악할 수 있는 멋진 워드클라우드를 만들어 봅시다.

```python
file = open('news.txt',"r",encoding='UTF8') #뉴스 기사를 가져온다.
text = file.read()
file.close()

wordList = text.split() #단어 단위로 잘라 리스트에 담는다.
```

```python
wordList
```

```python
worduniq = set(wordList) #중복된 값을 없애주기 위해

wordCount = {} #단어와 빈도수를 담을 빈 딕셔너리 공간

for w in worduniq:
    wordCount[w] = wordList.count(w)
    
del_word = ['a','is','the','of','on','it','from','an','more','at','its','been'.'just','have','this','from','and','are','has','by','for','in','to','was','The','than','that','be','as']
for w in del_word :
    if w in wordCount :
        del wordCount[w]
```

```python
wordCount
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width=1000, height=600).generate_from_frequencies(wordCount)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.show()
```
